```python
# create 

import numpy as np
data_pair_drop = []
for index in data_corr.index:
    for col in data_corr.columns:
        if data_corr.loc[index,col] >0.8 and index!=col and (index,col) not in data_pair_drop:
            data_pair_drop.append((index,col))
new_datav2 = np.unique([col[1] for col in data_pair_drop])   
```

```python
#x = data[[c for c in data.columns if c !='Churn' and c !='gender' ]]
x = data[['SeniorCitizen','tenure','MonthlyCharges']]
y = data['Churn'].map({'Yes':1,'No':0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.pipeline import Pipeline    # 导入sklearn中的Pipeline
from sklearn.preprocessing import StandardScaler   # 将数据均值移除，变为均值为0，方差为1的数据
from sklearn.decomposition import PCA  
import xgboost as xgb
model = xgb.XGBClassifier()

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('model', model)
])
pipeline.fit(x_train,y_train)
print(pipeline.score(x_test,y_test))
```
