### 实验操作步骤

#### 步骤1 导入绘图所需要的库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 步骤2 绘制正弦图

```python
# 使用numpy库生成数据,并使用matplotlib库中的plot方法对所生成的数据进行绘图
plt.style.use('seaborn-darkgrid')  # 选择图形主题
# 共有五个主题, darkgrid. whitegrid dark. white  ticks. 
# 在seaborn中设置主题使用set_style()
# 生成数据
x = np.arange(0,3*np.pi,0.1)
x
```

```python
y = np.sin(x)    # 取正弦值
y
```

```python
# matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data = None,**kargs):
# 绘制曲线图, *args中包含所需要的数据,曲线的颜色和样式表
plt.plot(x,y)   # 根据传入的数据,绘制图形
plt.show()
```

#### 绘制cos

```python
y1 = np.cos(x)
y1
```

```python
plt.plot(x,y1)
plt.show()
```

#### 步骤3 绘制子图

```python
plt.style.use('seaborn-whitegrid')
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sinx')
plt.subplot(2,1,2)
plt.plot(x, y_cos)
plt.title('Cosx')
plt.show()
```

```python
X = [1,2,3,4]
Y = [2,4,6,8]
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Result")
plt.style.use("seaborn-whitegrid")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('seaborn-dark')
plt.figure(figsize=(8,8))
plt.subplot(111)
X = np.linspace(-np.pi,np.pi,256,endpoint=True)
```

```python
C,S = np.cos(X),np.sin(X)
```

```python
plt.plot(X,C,color="blue",linewidth=1.0,label="Blue",linestyle="--")
```

```python
plt.plot(X,S, linewidth=1.0,label="Green",linestyle="-")
```

```python
plt.legend()
plt.xlim(-4.0,4.0)
plt.xticks(np.linspace(-4,4,9,endpoint=True))
plt.ylim(-1,1)
plt.xticks(np.linspace(-1,1,5,endpoint=True))
plt.show()
```
