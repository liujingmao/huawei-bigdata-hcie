```python
import pandas as pd
import numpy as np
data = pd.read_csv('dc_ppt.csv')
data.head()
```

```python
X = data.drop(['y'],axis=1)
y = data.loc[:,'y']
```

```python
from sklearn import tree
dc = tree.DecisionTreeClassifier(criterion='entropy')
dc.fit(X,y)
```

```python
%matplotlib inline
from pylab import *  
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(10,10))
mpl.rcParams['font.sans-serif'] = ['SimHei']
tree.plot_tree(dc,filled='True',feature_names=['动力', '能力', '兴趣', '时间'],class_names=['不合适','合适'])
```
